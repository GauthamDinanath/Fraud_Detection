```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 11, 5
```

```python
path='./creditcard.csv'
df=pd.read_csv(path)
df.head()
```

```python
print("Shape: ", df.shape)
df.describe()
```

```python
abnormal= df[df.Class==1]
normal  = df[df.Class==0]

print(f"Anomalies: {abnormal.shape}")
print(f"Normal:    {normal.shape}")
```

```python
#Analysis

class_counts = pd.value_counts(df['Class'], sort = True)
class_counts.plot(kind = 'bar', rot=0)
plt.title("Target Distribution")
plt.xticks(range(2), ["Normal", "Abnormal"])
plt.xlabel("Label")
plt.ylabel("Counts")
```

```python
print("\n\tAbnormal")
print(abnormal.Amount.describe())
print("\n\tNormal")
print(normal.Amount.describe())
```

```python
def plot_histogram(df, bins, column, log_scale=False):
    bins = 100
    abnormal = df[df.Class == 1]
    normal = df[df.Class == 0]
    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
    fig.suptitle(f'Counts of {column} by Target Class')
    ax1.hist(abnormal[column], bins = bins, color="red")
    ax1.set_title('Abnormal')
    ax2.hist(normal[column], bins = bins, color="orange")
    ax2.set_title('Normal')
    plt.xlabel(f'{column}')
    plt.ylabel('Count')
    if log_scale:
        plt.yscale('log')
    plt.xlim((np.min(df[column]), np.max(df[column])))
    plt.show()
    
def plot_scatter(df, x_col, y_col, sharey = False):
    anomalies = df[df.Class == 1]
    normal = df[df.Class == 0]
    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, sharey=sharey)
    fig.suptitle(f'{y_col} over {x_col} by Target Class')
    ax1.scatter(anomalies[x_col], anomalies[y_col], color="red")
    ax1.set_title('Anomaly')
    ax2.scatter(normal[x_col], normal[y_col], color="orange")
    ax2.set_title('Normal')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.show()
```

##  Amount vs Target class Relationship

```python
plt.scatter(df.Amount, df.Class)
plt.title("Transaction Amounts by Target")
plt.ylabel("Target")
plt.yticks(range(2), ["Normal", "Abnormal"])
plt.xlabel("Transaction Amounts ($)")
plt.show()
```

```python
bins = 100
plot_histogram(df, bins, "Amount", log_scale=True)
```

```python
plt.hist(abnormal.Amount, bins = bins, color="red")
plt.show()
```

##  Amount vs Time Relationship

```python
plt.scatter(df.Time, df.Class)
plt.title("Transactions over Time by Class")
plt.ylabel("Class")
plt.yticks(range(2), ["Normal", "Anomaly"])
plt.xlabel("Time (s)")
plt.show()
```

```python
plot_scatter(df, "Time", "Amount")
```

```python
plot_histogram(df, bins, "Time",log_scale=True)
```

##  V vs (Amount and Time) Relationship

```python
for f in range(1, 29):
    print(f'V{f} Counts')
    plot_histogram(df, bins, f'V{f}')
```

```python
for f in range(1, 29):
    print(f'V{f} vs Time')
    plot_scatter(df, "Time", f'V{f}', sharey=True)
```

```python
for f in range(1, 29):
    print(f'Amount vs V{f}')
    plot_scatter(df,  f'V{f}', "Amount", sharey=True)
```
